##How to preprocess annotated data to spaCy V3 format

```python
import pandas as pd
import spacy
from spacy.tokens import DocBin
import ast
```

```python
print(spacy.__version__)
```

```python
df = pd.read_csv('/content/01_03_data_annotation_for_named_entity_recognition.csv')
```

```python
df.head()
```

```python
df['annotations'] = df['annotations'].apply(ast.literal_eval)
```

```python
df
```

```python
df.info()
```

```python
N = 100
df = pd.concat([df] * N, ignore_index=True)
```

```python
df.info()
```

```python
train_data = df['annotations'][:1200].tolist()
test_data = df['annotations'][1200:].tolist()
```

```python
nlp = spacy.blank("en") # load a new spacy model
db = DocBin() # create a DocBin object
```

```python
for text, annot in train_data: # data in spacy tuple
        doc = nlp.make_doc(text) # create doc object from text
        ents = []
        for start, end, label in annot["entities"]: # add character indexes
            span = doc.char_span(start, end, label=label, alignment_mode="contract")
            if span is None:
                print("Skipping entity")
            else:
                ents.append(span)
```

```python
doc.ents = ents # label the text with the ents
db.add(doc)

db.to_disk("./train.spacy") # save the docbin object
```

```python
nlp2 = spacy.blank("en")
db2 = DocBin()
```

```python
for text, annot in test_data: # data in spacy tuple
        doc2 = nlp2.make_doc(text) # create doc object from text
        ents2 = []
        for start, end, label in annot["entities"]: # add character indexes
            span = doc2.char_span(start, end, label=label, alignment_mode="contract")
            if span is None:
                print("Skipping entity")
            else:
                ents2.append(span)
```

```python
doc2.ents = ents2 # label the text with the ents
db2.add(doc2)

db2.to_disk("./test.spacy") # save the docbin object
```
